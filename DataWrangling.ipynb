#### Karena data "hour.csv" Lebih baik, sebaiknya menggunakan kumpulan data "hour.csv" karena datanya lebih banyak dan lebih detail dibandingkan data "day.csv"

```python
# Mengganti nama dataset jam menjadi sewa_cuaca_waktu
sewa_cuaca_waktu = jam
# Mengubah tipe data kolom dteday menjadi datetime
sewa_cuaca_waktu['dteday'] = pd.to_datetime(sewa_cuaca_waktu['dteday'])
# Menampilkan tipe data untuk dataset gabungan
print('Tipe data untuk dataset gabungan: ', sewa_cuaca_waktu.dtypes)
```

```python
# Mengubah nama kolom dalam dataset yang digabungkan agar lebih mudah dibaca
sewa_cuaca_waktu.rename(columns={'dteday':'tanggal', 
                                 'yr':'tahun', 
                                 'mnth':'bulan', 
                                 'hr':'jam', 
                                 'season':'musim',
                                 'holiday':'libur',
                                 'weathersit':'kondisi_cuaca', 
                                 'casual':'pengguna_casual', 
                                 'registered':'pengguna_terdaftar', 
                                 'hum':'kelembaban', 'cnt':'jumlah',
                                 'weekday':'hari',
                                 'atemp':'suhu_terasa',
                                 'workingday':'hari_kerja'
                                 }, inplace=True)

# Memeriksa nama-nama kolom untuk dataset yang digabungkan
print('Nama kolom untuk dataset yang digabungkan: ', sewa_cuaca_waktu.columns)
```

```python
# Mengubah nama dan nilai dalam kolom agar lebih mudah dibaca


# Mengubah nilai kolom musim menjadi nama-nama musim
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['musim'] == 1, 'musim'] = 'Dingin'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['musim'] == 2, 'musim'] = 'Semi'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['musim'] == 3, 'musim'] = 'Panas'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['musim'] == 4, 'musim'] = 'Gugur'

# Mengubah nilai kolom kondisi_cuaca menjadi nama-nama kondisi cuaca
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['kondisi_cuaca'] == 1, 'kondisi_cuaca'] = 'Cerah'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['kondisi_cuaca'] == 2, 'kondisi_cuaca'] = 'Berawan'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['kondisi_cuaca'] == 3, 'kondisi_cuaca'] = 'Hujan Ringan'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['kondisi_cuaca'] == 4, 'kondisi_cuaca'] = 'Hujan Lebat'

# Kolom suhu: normalisasi suhu dalam derajat Celsius. Rumus: (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (dari deskripsi dataset)
t_min = -8  # Nilai suhu minimum dalam derajat Celsius
t_max = 39  # Nilai suhu maksimum dalam derajat Celsius

# Mengonversi nilai suhu yang dinormalisasi kembali ke dalam derajat Celsius
sewa_cuaca_waktu['temp'] = sewa_cuaca_waktu['temp'].apply(lambda x: (x*(t_max-t_min))+t_min)

# Menampilkan rentang nilai untuk kolom suhu
print('Rentang nilai untuk kolom suhu: ', sewa_cuaca_waktu['temp'].min(), sewa_cuaca_waktu['temp'].max())

# Kolom suhu_terasa: normalisasi suhu terasa dalam derajat Celsius. Rumus: (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (dari deskripsi dataset)
at_min = -16
at_max = +50

# Mengonversi nilai suhu terasa yang dinormalisasi kembali ke dalam derajat Celsius
sewa_cuaca_waktu['suhu_terasa'] = sewa_cuaca_waktu['suhu_terasa'].apply(lambda x: (x*(at_max-at_min))+at_min)

# Menampilkan rentang nilai untuk kolom suhu_terasa
print('Rentang nilai untuk kolom suhu_terasa: ', sewa_cuaca_waktu['suhu_terasa'].min(), sewa_cuaca_waktu['suhu_terasa'].max())

# Kolom kelembaban: normalisasi kelembaban. Rumus: (kelembaban-kelembaban_min)/(kelembaban_max-kelembaban_min), kelembaban_min=0, kelembaban_max=1 (dari deskripsi dataset)
sewa_cuaca_waktu['kelembaban'] = sewa_cuaca_waktu['kelembaban'].apply(lambda x: x/100)

# Kolom kecepatan_angin: Kecepatan angin yang dinormalisasi. Nilainya dibagi dengan 67 (maksimum) (dari deskripsi dataset)
kecepatan_min = 0
kecepatan_max = 0.85

# Mengonversi nilai kecepatan_angin yang dinormalisasi kembali ke dalam nilai yang tidak dinormalisasi (km/jam) Nilainya dibagi dengan 67 (dari deskripsi dataset)       
sewa_cuaca_waktu['windspeed'] = sewa_cuaca_waktu['windspeed'].apply(lambda x: (x * 67))

# Menampilkan rentang nilai untuk kolom kecepatan_angin
print('Rentang nilai untuk kolom kecepatan_angin: ', sewa_cuaca_waktu['windspeed'].min(), sewa_cuaca_waktu['windspeed'].max())

# Mengubah nilai kolom kelembaban menjadi persentase
sewa_cuaca_waktu['kelembaban'] = sewa_cuaca_waktu['kelembaban'].apply(lambda x: x*100)

# Menampilkan rentang nilai untuk kolom kelembaban
print('Rentang nilai untuk kolom kelembaban: ', sewa_cuaca_waktu['kelembaban'].min(), sewa_cuaca_waktu['kelembaban'].max())

# Mengubah nilai kolom bulan menjadi nama-nama bulan
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 1, 'bulan'] = 'Januari'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 2, 'bulan'] = 'Februari'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 3, 'bulan'] = 'Maret'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 4, 'bulan'] = 'April'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 5, 'bulan'] = 'Mei'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 6, 'bulan'] = 'Juni'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 7, 'bulan'] = 'Juli'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 8, 'bulan'] = 'Agustus'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 9, 'bulan'] = 'September'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 10, 'bulan'] = 'Oktober'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 11, 'bulan'] = 'November'
sewa_cuaca_waktu.loc[sewa_cuaca_waktu['bulan'] == 12, 'bulan'] = 'Desember'

# Menampilkan nilai unik untuk setiap kolom dalam dataset sewa_cuaca_waktu
print('Nilai unik untuk setiap kolom dalam dataset sewa_cuaca_waktu: ', sewa_cuaca_waktu.nunique())

# Menampilkan 5 baris pertama dari dataset sewa_cuaca_waktu untuk memeriksa perubahan yang telah dilakukan pada nama kolom dan nilai
display(sewa_cuaca_waktu.head())
```

```python
# Membuat kolom baru bernama 'periode_hari' untuk mengklasifikasikan jam dalam dataset sewa_cuaca_waktu
# Membuat fungsi untuk mengklasifikasikan jam
def periode_hari(jam):
    if jam >= 0 and jam <= 6:
        return 'Malam'
    elif jam > 6 and jam <= 12:
        return 'Pagi'
    elif jam > 12 and jam <= 18:
        return 'Sore'
    else:
        return 'Malam'

# Menggunakan fungsi tersebut pada kolom jam
sewa_cuaca_waktu['periode_hari'] = sewa_cuaca_waktu['jam'].apply(periode_hari)
```

```python
# Membuat kolom baru bernama 'kehangatan' untuk mengklasifikasikan suhu terasa dalam dataset sewa_cuaca_waktu
# Membuat fungsi untuk mengklasifikasikan suhu terasa
def kehangatan(suhu_terasa):
    if suhu_terasa <= 12:
        return 'Dingin'
    elif suhu_terasa > 12 and suhu_terasa <= 20:
        return 'Hangat'
    elif suhu_terasa > 20 and suhu_terasa <= 30:
        return 'Panas'
    else:
        return 'Sangat Panas'
    
# Menggunakan fungsi tersebut pada kolom suhu_terasa
sewa_cuaca_waktu['kehangatan'] = sewa_cuaca_waktu['suhu_terasa'].apply(kehangatan)

# Menampilkan nilai-nilai dari kolom kehangatan
print('Nilai dari kolom kehangatan: ', sewa_cuaca_waktu['kehangatan'].unique())
```

```python
# Membuat kolom baru bernama 'tingkat_kelembaban' untuk mengklasifikasikan kelembaban dalam dataset sewa_cuaca_waktu
# Membuat fungsi untuk mengklasifikasikan kelembaban
def tingkat_kelembaban(kelembaban):
    if kelembaban <= 0.40:
        return 'Rendah'
    elif kelembaban > 0.40 and kelembaban <= 0.70:
        return 'Sedang'
    else:
        return 'Tinggi'
    
# Menggunakan fungsi tersebut pada kolom kelembaban
sewa_cuaca_waktu['tingkat_kelembaban'] = sewa_cuaca_waktu['kelembaban'].apply(tingkat_kelembaban)

# Menampilkan nilai-nilai dari kolom tingkat_kelembaban
print('Nilai dari kolom tingkat_kelembaban: ', sewa_cuaca_waktu['tingkat_kelembaban'].unique())
```
